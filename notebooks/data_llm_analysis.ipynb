```python
import os
import json
import requests

origin_data_folder="/home/thhuang/voc_data/cht_sales_demo/data"
processed_data_folder="/home/thhuang/myproject/langflow_implement/processed_json"

langflow_api="http://localhost:7860/api/v1/run/75239f90-969d-4452-b3ce-b256cce9ba2d?stream=false"
# langflow_api="http://localhost:7860/api/v1/run/3609db05-62d9-4cec-8dd6-5f9a2fdde803?stream=false"
```

```python
#讀取origin_data_folder的json檔案
def read_json_files(folder_path):
    json_data = []
    for filename in os.listdir(folder_path):
        if filename.endswith('.json'):
            file_path = os.path.join(folder_path, filename)
            with open(file_path, 'r') as file:
                data = json.load(file)
                json_data.append(data)
    return json_data

#將json檔案寫入processed_data_folder
def write_json_files(folder_path, json_data):
    for i, data in enumerate(json_data):
        file_path = os.path.join(folder_path, f'processed_{i}.json')
        with open(file_path, 'w') as file:
            json.dump(data, file, indent=4,ensure_ascii=False)
```

```python
data_list=read_json_files(origin_data_folder)
```

```python
# 準備 API 請求
headers = {"Content-Type": "application/json"}

result_list=[]

total_dialogues=len(data_list)
dialogues_counter=1


for i in data_list:
    
    counter=1

    total_sentences=len(i["conversations"])
    # processed_conversations=[]
    for sentence in i["conversations"]:
        text=[]
        for sentence in i["conversations"]:
            if sentence["type"]=="agent":
                text.append(f"agent: {sentence["content"]}")
            elif sentence["type"]=="customer":
                text.append(f"customer: {sentence["content"]}")
        # print(text)
        data={"input_value": json.dumps(text)}

    try:
        response=requests.post(langflow_api, headers=headers, json=data)
        content=response.json()['outputs'][0]['outputs'][0]['results']['text']["text"]

        summary=[]
        summary.append(f"customer: {json.loads(content)["customer"]}")
        summary.append(f"agent: {json.loads(content)["agent"]}")
        # print(summary)
        
        i["summary"]=summary
        print(f"對話處理進度：{dialogues_counter}/{total_dialogues}")
    except:
        print("Error in response:", response.text)
    

    result_list.append(i)

    file_path = os.path.join(processed_data_folder, f'processed_{i["contactid"]}.json')
    with open(file_path, 'w') as file:
        json.dump(i, file, indent=4,ensure_ascii=False)


    dialogues_counter+=1



```
